```python
from refractio import snowflake
```

```python
dataset = snowflake.get_dataframe("PHARMA_BRAND")
```

```python
dataset
```

```python
import matplotlib.pyplot as plt #import pandas and matplotlib
import pandas as pd
alphatumab_data = dataset[dataset['BRAND']=='ALPHATUMAB'] #to read the dataset where brand is ALPHATUMAB
alphatumab_data.loc[:,'DATE'] = pd.to_datetime(alphatumab_data.loc[:,'DATE']) #To DATETIME format conversion for python to understand
monthly_data = alphatumab_data.resample('M', on='DATE').sum() #Aggregation of data to Monthly
plt.figure(figsize = (12,5)) #visualisation
plt.plot(monthly_data.index , monthly_data['TRX'],  marker='o',  linestyle = '-')
plt.title('Trend of TRx over time for ALPHATUMAB')
plt.xlabel('DATE')
plt.ylabel('No of TRx')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
betatab_data = dataset[dataset['BRAND']=='BETATAB']
betatab_data['DATE'] = pd.to_datetime(betatab_data['DATE'])
monthly_data = betatab_data.resample('M', on='DATE').sum()
plt.figure(figsize = (15,5))
plt.plot(monthly_data.index , monthly_data['TRX'],  marker='o',  linestyle = '-')
plt.title('Trend of TRx over time for BETATAB')
plt.xlabel('Date')
plt.ylabel('No of TRx')
plt.grid(True)
plt.show()
```
